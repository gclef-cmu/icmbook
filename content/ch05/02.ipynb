# 5.2 The phasor

Now we bring the complex plane back into a sound context. Recall the basic sinusoid, the most elementary periodic sound. In Chapter 1 we wrote it as $a\sin(\omega t + \phi)$, but here we will use the cosine form

$$x(t) = a\cos(\omega t)$$

for reasons that will become clear momentarily. These forms are interchangeable: $\cos(\omega t) = \sin(\omega t + \pi/2)$, so switching to cosine just fixes a particular initial phase.

:::{tip}
Here $\omega$ is angular frequency, in units of ${unit}`radians,second`$. If you're rusty on angular frequency, revisit {ref}`sec-angular-frequency`. We will use angular frequency regularly from here on, since it spares us from writing $2\pi f$ everywhere.
:::

Suppose we want to transform this basic sinusoid so that it operates in the complex plane. How might we do that? We apply Euler's formula. Starting from $e^{j\theta} = \cos\theta + j\sin\theta$, we multiply both sides by $a$ and then let the angle vary with time as $\theta(t) = \omega t$:

$$
\begin{aligned}
e^{j\theta} &= \cos\theta + j\sin\theta \\
a\, e^{j\theta} &= a\cos\theta + j\, a\sin\theta \\
a\, e^{j\omega t} &= \underbrace{a\cos(\omega t)}_{\text{basic sinusoid (real)}} + \underbrace{j\, a\sin(\omega t)}_{\text{analytical tool (imaginary)}}.
\end{aligned}
$$

What have we accomplished? Two things:

1. We brought our basic sinusoid into the complex plane by complementing it with an _analytical tool_, the term $j\, a\sin(\omega t)$. Where the basic sinusoid is always real-valued, this tool is always imaginary-valued, and it is always exactly $\pi/2$ radians out of phase with the basic sinusoid.
1. We used Euler's formula to fold the two real sinusoids into a single, compact {vocab}`complex sinusoid`, $a\, e^{j\omega t}$.

What does a complex sinusoid look like? It is a vector of length $a$ that rotates counterclockwise, tracing the outline of a circle of radius $a$ in the complex plane. As it rotates, its real part traces a cosine and its imaginary part traces a sine:

:::{figure}
![Left: a vector of length 1 in the complex plane pointing to a dot on the unit circle, with dashed lines projecting onto the real and imaginary axes. Middle: a cosine wave over one cycle, the real projection. Right: a sine wave over one cycle, the imaginary projection.](./assets/fig-phasor.png)

A complex sinusoid, or _phasor_, $a\, e^{j\omega t}$ at one instant (here $a = 1$). Left: in the complex plane it is a rotating vector. Middle and right: its real and imaginary parts, projected out over time, are a cosine and a sine. The phasor completes one full rotation every $1/f$ seconds, where $f = \omega / 2\pi$.
:::

A complex sinusoid is very commonly called a {vocab}`phasor`. Plainly, a phasor is just a fancy way to draw a circle over and over. We can define it as a function of time, parameterized by its amplitude $a$ and angular frequency $\omega$:

$$\text{phasor}_{a, \omega}(t) = a\, e^{j\omega t} : \mathbb{R} \to \mathbb{C}.$$

The subscripts $a$ and $\omega$ are fixed parameters that pick out _which_ phasor we mean, exactly as $a$ and $f$ parameterize the basic sinusoid. The lone input to the function is still time $t$.

:::{important}
Like the basic sinusoid, a phasor is a function of _time_, not of frequency. The only difference is that it rotates in the complex plane rather than oscillating along a single real axis.
:::

```python
# hide
# no-output
# VS Code-only preview (plotly + ipywidgets). Skipped under the book build:
# every ipywidgets model instantiated at build time gets baked into the page
# as ~5 MB of inert widget-state JSON, so the build sets ICM_BOOK_BUILD=1
# (Makefile) and this cell builds nothing there. Run it locally as usual.
import os

if not os.environ.get("ICM_BOOK_BUILD"):
    import numpy as np
    import plotly.graph_objects as go
    from plotly.subplots import make_subplots
    import ipywidgets as widgets
    from IPython.display import display

    # time axis for the waveform
    t = np.linspace(0, 2, 800)

    # make the figure
    fig = go.FigureWidget(make_subplots(rows=1, cols=2, horizontal_spacing=0.2))

    # rotating vector
    fig.add_scatter(name = "Complex Vector", x=[0, 1], y=[0, 0], mode="lines+markers", row=1, col=1)

    # waveform
    fig.add_scatter(name = "Waveform", x=t, y=np.sin(2*np.pi*t), mode="lines", row=1, col=2)

    # moving point
    fig.add_scatter(name = "Reference Point", x=[0], y=[0], mode="markers", row=1, col=2)

    # fix the axes
    fig.update_xaxes(range=[-2, 2], scaleanchor="y", scaleratio=1, row=1, col=1, title="Re")
    fig.update_yaxes(range=[-2, 2], scaleanchor="x", scaleratio=1, row=1, col=1, title="Im")
    fig.update_xaxes(range=[0, 2], row=1, col=2, title="Time (s)")
    fig.update_yaxes(range=[-2, 2], row=1, col=2, title="Amplitude")
    # make it look nicer
    fig.update_layout(
        font={"family": "Times New Roman", "size": 10},
        margin={"pad" : 10}
    )

    # sliders
    fullWidth = style = {'description_width': 'initial'}
    amp = widgets.FloatSlider(description="Amplitude",min=0,max=2,value=1,step=0.05, style=fullWidth)
    freq = widgets.FloatSlider(description="Frequency (Hz)",min=0.25,max=5,value=1,step=0.05, style=fullWidth)
    time = widgets.IntSlider(description="Time (ms)",min=0,max=2000,value=0,step=1, style=fullWidth)
    # make the time move automatically with a play button
    play = widgets.Play(value=0,min=0,max=2000, step=10, interval=30)

    # make play move the time slider
    widgets.jslink((play, "value"), (time, "value"))

    # redraw the figure whenever anything changes
    def update(time, amp, freq):
        time /= 1000
        angle = 2*np.pi*freq*time

        # rotating vector
        fig.data[0].x = [0, amp*np.cos(angle)]
        fig.data[0].y = [0, amp*np.sin(angle)]

        # waveform
        fig.data[1].y = amp*np.sin(2*np.pi*freq*t)

        # moving point
        fig.data[2].x = [time]
        fig.data[2].y = [amp*np.sin(angle)]

    widgets.interactive_output(
        update,
        {
            "time": time,
            "amp": amp,
            "freq": freq,
        },
    )

    display(widgets.VBox([ amp, freq, widgets.HBox([play, time])]), fig)
```

```python
# hide
# --- The page widget: the phasor with amplitude & frequency sliders ----------
# Self-contained: everything this widget needs is in this one cell. Built
# with the book's icm_widgets ParamPlayer (see the interactive template):
# matplotlib renders every slider combination at build time, and the page
# runtime just swaps the pre-rendered frames — sliders with no kernel and no
# HTML in this cell. (icm_widgets ships with the book: pre-installed in the
# build environment and in the browser kernel.)
import numpy as np
import matplotlib.pyplot as plt
from myst_nb import glue
from icm_widgets import ParamPlayer, slider, play

plt.rcParams.update({
    "figure.dpi": 64,                # sets the widget's on-page size; frames
    "text.usetex": False,            # are SVG, crisp at any display density
    "font.size": 12,
    "axes.spines.top": False, "axes.spines.right": False,
    "axes.edgecolor": "#6D6E71",
    "xtick.color": "#6D6E71", "ytick.color": "#6D6E71",
    "xtick.labelcolor": "#3b3b3b", "ytick.labelcolor": "#3b3b3b",
    "axes.labelcolor": "#3b3b3b",
})

fig, (ax_c, ax_w) = plt.subplots(
    1, 2, figsize=(8.2, 3.8),
    gridspec_kw={"width_ratios": [1, 1.6], "wspace": 0.35},
)
fig.subplots_adjust(bottom=0.16)
ax_c.set_xlim(-2, 2); ax_c.set_ylim(-2, 2)
ax_c.set_aspect("equal", "box")
ax_c.set_xlabel("Re"); ax_c.set_ylabel("Im")
ax_w.set_xlim(0, 2); ax_w.set_ylim(-2, 2)
ax_w.set_xlabel("Time (s)"); ax_w.set_ylabel("Amplitude")

# dynamic artists, moved by update() — the FuncAnimation discipline
t = np.linspace(0, 2, 800)            # waveform time axis (two seconds)
theta = np.linspace(0, 2 * np.pi, 220)
circle, = ax_c.plot([], [], color="#E0E0E0", lw=1.5)   # the path the tip traces
vec,    = ax_c.plot([], [], color="#C41230", lw=2.2, marker="o", markevery=[1])
wave,   = ax_w.plot([], [], color="#007BC0", lw=1.8)
dot,    = ax_w.plot([], [], "o", color="#C41230")


def update(amplitude, frequency, time):
    a, f = amplitude, frequency
    phase = 2 * np.pi * f * time         # the angle the vector has reached
    circle.set_data(a * np.cos(theta), a * np.sin(theta))
    vec.set_data([0, a * np.cos(phase)], [0, a * np.sin(phase)])
    wave.set_data(t, a * np.sin(2 * np.pi * f * t))
    dot.set_data([time], [a * np.sin(2 * np.pi * f * time)])


widget = ParamPlayer(
    fig, update, dynamic=[circle, vec, wave, dot],
    # Per-artist parameter dependence: each group renders as its own layer,
    # so the low-ink rotating marks (vector + dot) can afford 48 played time
    # frames while the waveform layer stays at the 8x8 slider grid.
    depends={circle: ("amplitude",), wave: ("amplitude", "frequency"),
             vec: ("amplitude", "frequency", "time"),
             dot: ("amplitude", "frequency", "time")},
    amplitude=slider(0.25, 2.0, step=0.25, value=1.0),
    frequency=slider(0.5, 4.0, step=0.5, value=1.0, unit="Hz"),
    # The played dimension is TIME across the whole two-second window, so the
    # dot sweeps every visible cycle (at f Hz the vector makes 2f turns per
    # sweep — endpoint excluded, so it loops seamlessly for every half-integer
    # frequency). rate= still makes a 2 Hz phasor spin twice as fast.
    # 96 frames over the 4 s sweep = 24 fps at the 1 Hz default — the house
    # motion rate; the runtime paces by wall clock, so frames buy smoothness.
    time=play(0, 2.0, frames=96, period=4.0, rate="frequency"),
)

# Register for the book page (embed with `{glue}`phasor-rotation``); the
# trailing expression bakes the widget into the page.
glue("phasor-rotation", widget, display=False)
widget
```

Take time to study this. Deriving the phasor is the main reason we reviewed the complex plane. **The complex sinusoid is perhaps the single most important expression in computer music.** It captures the periodic essence of sound in the basic sinusoid, and, as we will now see, it gives rise to the Fourier transform that uncovers a unique sinusoidal recipe for any sound.
