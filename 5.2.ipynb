```python
import numpy as np
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import ipywidgets as widgets
from IPython.display import display

# time axis for the waveform
t = np.linspace(0, 2, 800)

# make the figure
fig = go.FigureWidget(make_subplots(rows=1, cols=2))

# rotating vector
fig.add_scatter(name = "Complex Vector", x=[0, 1], y=[0, 0], mode="lines+markers", row=1, col=1)

# waveform
fig.add_scatter(name = "Waveform", x=t, y=np.sin(2*np.pi*t), mode="lines", row=1, col=2)

# moving point
fig.add_scatter(name = "Moving Point", x=[0], y=[0], mode="markers", row=1, col=2)

# fix the axes
fig.update_xaxes(range=[-2, 2], scaleanchor="y", scaleratio=1, row=1, col=1)
fig.update_yaxes(range=[-2, 2], scaleanchor="x", scaleratio=1, row=1, col=1)
fig.update_xaxes(range=[0, 2], row=1, col=2)
fig.update_yaxes(range=[-2, 2], row=1, col=2)

# sliders
amp = widgets.FloatSlider(description="Amplitude",min=0,max=2,value=1,step=0.05)
freq = widgets.FloatSlider(description="Frequency (Hz)",min=0.25,max=5,value=1,step=0.05)
time = widgets.IntSlider(description="Time (ms)",min=0,max=2000,value=0,step=1)
# make the time move automatically with a play button
play = widgets.Play(value=0,min=0,max=2000, step=20)

# make play move the time slider
widgets.jslink((play, "value"), (time, "value"))

# redraw the figure whenever anything changes
def update(time, amp, freq):
    time /= 1000
    angle = 2*np.pi*freq*time

    # rotating vector
    fig.data[0].x = [0, amp*np.cos(angle)]
    fig.data[0].y = [0, amp*np.sin(angle)]

    # waveform
    fig.data[1].y = amp*np.sin(2*np.pi*freq*t)

    # moving point
    fig.data[2].x = [time]
    fig.data[2].y = [amp*np.sin(angle)]

widgets.interactive_output(
    update,
    {
        "time": time,
        "amp": amp,
        "freq": freq,
    },
)

display(widgets.VBox([widgets.HBox([play, time]), amp, freq]), fig)
```
